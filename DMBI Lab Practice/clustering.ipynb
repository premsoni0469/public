```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, max_iter=100, init="k-means++", n_init=10, random_state=101)
```

```python
y_kmeans = kmeans.fit_predict(X)
```

```python
print(kmeans.cluster_centers_)
```

```python
X_np = X.values
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X_np[y_kmeans == 0, 0], X_np[y_kmeans == 0, 1], c='red', s=100, label='setosa')
plt.scatter(X_np[y_kmeans == 1, 0], X_np[y_kmeans == 1, 1], c='blue', s=100, label='versicolour')
plt.scatter(X_np[y_kmeans == 2, 0], X_np[y_kmeans == 2, 1], c='green', s=100, label='virginica')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1],  c='orange', s=100, label='cluster centers')

plt.legend()
plt.show()
```
