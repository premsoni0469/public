```python
max = df['Quantity'].max()
min = df['Quantity'].min()

print(max, min)
```

```python
print(df['Quantity'].mean())
print(df['Quantity'].median())
print(df['Quantity'].mode())
print(df['Quantity'].std())
```

```python
Q1 = df['Quantity'].quantile(0.25)
Q3 = df['Quantity'].quantile(0.75)

IQR = Q3 - Q1
 
print("Q1: ", Q1)
print("Q2: ", df['Quantity'].median())
print("Q3: ", Q3)
print("IQR: ", IQR)
```

```python
whisker_width = 1.5
low_out = df[df['Quantity'] < Q1 - whisker_width * IQR]
print("low_out: ", low_out)
high_out = df[df['Quantity'] > Q3 + whisker_width * IQR]
print("high_out: ", high_out)
```
